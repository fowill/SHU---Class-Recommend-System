```python
names = ['student_id']
for i in range(10):
    part = ['lesson_id'+str(i+1),'lesson_name'+str(i+1),'gpa'+str(i+1),'recommend_score'+str(i+1)]
    names += part
chooseDF = pd.read_csv('学生-选课.csv',index_col=False,encoding='gbk',names=names)
chooseDF = chooseDF.fillna('None')
```

```python
c_lessonDF = lessonDF[['lesson_id','lesson_type','avg_gpa','var_gpa','avg_recommend_score','var_recommend_score']]
```

```python
c_lessonDF
```

```python
#lessonDF，处理离散的features

categorical_features = ['lesson_type']
c_lessonDF_cat = pd.get_dummies(c_lessonDF[categorical_features])
c_lessonDF = c_lessonDF.drop(categorical_features, axis=1)
c_lessonDF = pd.concat([c_lessonDF, c_lessonDF_cat], axis=1)

c_lessonDF
```

```python
from sklearn.preprocessing import StandardScaler
stds = StandardScaler()
knn_lessonDF = c_lessonDF
knn_lessonDF.iloc[:,1:5] = stds.fit_transform(c_lessonDF.iloc[:,1:5])
```

```python
knn_lessonDF
```

```python
#SSE 样本距离最近的聚类中心的距离总和 （簇内误差平方和）

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

distortions = []
for i in range(1, 40):
        kmeans = KMeans(n_clusters=i, random_state=0).fit(knn_lessonDF.iloc[:,1:])
        distortions.append(kmeans.inertia_)

plt.plot(range(1,40), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()

```

```python
#轮廓系数

from sklearn import metrics
import matplotlib.pyplot as plt

scores = []
for i in range(2, 40):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(knn_lessonDF.iloc[:,1:])
    scores.append(metrics.silhouette_score(knn_lessonDF.iloc[:,1:], kmeans.labels_ , metric='euclidean'))

plt.plot(range(2, 40), scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('silhouette_score')
plt.show()

```

```python
#Calinski-Harabaz Index

from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
ch_scores = []
for i in range(2, 40):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(knn_lessonDF.iloc[:,1:])
    ch_scores.append(metrics.calinski_harabasz_score(knn_lessonDF.iloc[:,1:], kmeans.labels_))

plt.plot(range(2, 40), ch_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('calinski_harabaz_score')
plt.show()

```

```python
#取n=20

kmeans = KMeans(n_clusters=20, random_state=0).fit(knn_lessonDF.iloc[:,1:])
knn_lessonDF['label'] = kmeans.labels_

knn_lessonDF
```

```python
#计算每个同学在每个类别课程里的平均得分、平均打分

student_gpa_vectors = []
student_recommend_score_vectors = []

for student_id in range(15000):
    personal_gpa_vector = [0]*20
    personal_recommend_score_vector = [0]*20
    personal_count_vector = [0]*20
    for n_class in range(10):
        lesson_id = chooseDF.loc[student_id,'lesson_id'+str(n_class+1)]
        lesson_gpa = chooseDF.loc[student_id,'gpa'+str(n_class+1)]
        lesson_recommend_score = chooseDF.loc[student_id,'recommend_score'+str(n_class+1)]
        #print(lesson_id,lesson_gpa,lesson_recommend_score)
        if lesson_id != 'None':
            lesson_type = knn_lessonDF.loc[lesson_id,'label']
            personal_count_vector[lesson_type] += 1
            personal_gpa_vector[lesson_type] += lesson_gpa
            personal_recommend_score_vector[lesson_type] += lesson_recommend_score
    #print('personal_count_vector',personal_count_vector)
    #print('personal_gpa_vector',personal_gpa_vector)
    #print('personal_recommend_score_vector',personal_recommend_score_vector)
    
    for i in range(20):
        if personal_count_vector[i] > 0:
            personal_gpa_vector[i] /= personal_count_vector[i]
            personal_recommend_score_vector[i] /= personal_count_vector[i]

    student_gpa_vectors.append(personal_gpa_vector)
    student_recommend_score_vectors.append(personal_recommend_score_vector)
        
    
```

```python
arr1 = np.array(student_gpa_vectors)
arr2 = np.array(student_recommend_score_vectors)
arr = np.hstack((arr1,arr2))
```

```python
arr
```

```python
student_vector_DF = pd.DataFrame(arr)
student_vector_DF
```

```python
student_vector_DF.values
```

```python
#对于每一个学生，计算每个其他学生与他的距离

def eucliDist(A,B):
    return np.sqrt(sum(np.power((A - B), 2)))

student_vector_array = student_vector_DF.values
student_distance_array = np.zeros([15000,15000])

#懒得做优化了
for i in range(15000):
    print(i)
    for j in range(15000):
        if i == j:
            d = 0
        else:
            d = eucliDist(student_vector_array[i],student_vector_array[j])
        student_distance_array[i,j] = d
```

```python
student_distance_array
```

```python
pd.DataFrame(student_distance_array).to_csv('distance.csv',header=None)
```

```python
#找出每个人最类似的100个人，并得到他们gpa最高、评分最高的10门课分别做推荐

import heapq

def change(x):
    if x == 0:
        return 1000
    return x

smallest100_distance_ls = []

for i in range(15000):
    distanceLs = [ change(x) for x in student_distance_array[i]]
    smallest100_id = heapq.nsmallest(100, range(len(distanceLs)), distanceLs.__getitem__)
    smallest100_distance_ls.append(smallest100_id)
    print(i)
```

```python
#读取studentDict

f = open('studentDict.txt','r')
a = f.read()
studentDict = eval(a)
f.close()
studentDict
```

```python
#整理对每个同学来说，100个最相似的同学，gpa平均得分最高or评价得分最高的10门课

def getLessonName(idLs):
    nameLs = []
    for lesson_id in idLs:
        nameLs.append(lessonDF.loc[lesson_id,'lesson_name'])
    return nameLs

resultDict = {}
for i in range(15000):
    resultDict[i] = {}
    
for i in range(15000):
    count_ls = [0] * 197
    gpa_ls = [0] * 197
    recommend_score_ls = [0] * 197
    for j in range(100):
        student_info = studentDict[smallest100_distance_ls[i][j]]
        chosed_class = student_info['chosed_class']
        chosed_class_gpa = student_info['chosed_class_gpa']
        chosed_class_recommend_score = student_info['chosed_class_recommend_score']
        for k in range(len(chosed_class)):
            count_ls[chosed_class[k]] += 1
            gpa_ls[chosed_class[k]] += chosed_class_gpa[k]
            recommend_score_ls[chosed_class[k]] += chosed_class_recommend_score[k]
    for n in range(197):
        if count_ls[n] > 0:
            gpa_ls[n] /= count_ls[n]
            recommend_score_ls[n] /= count_ls[n]
            
    resultDict[i]['gpa角度推荐'] = getLessonName(heapq.nlargest(10,range(len(gpa_ls)),gpa_ls.__getitem__))
    resultDict[i]['gpa角度得分'] = heapq.nlargest(10,gpa_ls) 
    resultDict[i]['满意角度推荐'] = getLessonName(heapq.nlargest(10,range(len(recommend_score_ls)),recommend_score_ls.__getitem__))
    resultDict[i]['满意角度得分'] = heapq.nlargest(10,recommend_score_ls)
```
